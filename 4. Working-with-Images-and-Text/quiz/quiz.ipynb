```python
from skimage.io import imread
import matplotlib.pyplot as plt
import skimage.morphology as skmorf
import collections
import numpy as np
from collections import Counter
import re
```

```python
hamburger = imread("http://lab.students.softuni.bg/assets/courseware/v1/692dbe85139f974150d80f57becfe093/asset-v1:SoftUni+DS101+June-2019+type@asset+block/hamburger.jpg")
```

```python
hamburger.shape
```

```python
hamburger.shape[0] * hamburger.shape[1] * hamburger.shape[2]
```

```python
red = hamburger[ :, :, 0]
green = hamburger[ :, :, 1]
blue = hamburger[ :, :, 2]
```

```python
red
```

```python
plt.hist(red.ravel(), bins = 256, color = "red")
plt.hist(green.ravel(), bins = 256, color = "green")
plt.hist(blue.ravel(), bins = 256, color = "blue")
plt.show()
```

```python
red.mean()
```

```python
green.mean()
```

```python
blue.mean()
```

```python
pic = skmorf.binary_opening(hamburger[ :, :, 2], selem=np.array([[1, 1, 1],[1, 1, 1],[1, 1, 1]]))
```

```python
plt.imshow(pic, cmap = "gray")
```

```python
pic
```

```python
pic = pic.astype(int)
```

```python
plt.imshow(pic)
```

```python
plt.hist(pic.ravel(), bins = 256)
plt.show()
```
